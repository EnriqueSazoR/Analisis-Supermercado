```python
# DataSet Número 2
ruta = r"C:\Users\Luis Sazo\Desktop\Ventas Supermercados\Data\annex2.csv"
DataSet2 = pd.read_csv(ruta)
DataSet2
```

```python
DataSet2.info()
DataSet2.isnull().sum()
# El DataSet no presenta nulos
```

```python
# Promedio de "Unit Selling Price"
PromedioVentaPorUnidad = DataSet2["Unit Selling Price (RMB/kg)"].mean()
print(f"El promedio de venta por unidad es ${round(PromedioVentaPorUnidad, 2)}")
```

```python
Ventas = DataSet2["Sale or Return"].value_counts().get("sale", 0)
Devoluciones = DataSet2["Sale or Return"].value_counts().get("return", 0)

print(f"Total de Ventas: {Ventas}")
print(f"Total de devoluciones: {Devoluciones}")
```
