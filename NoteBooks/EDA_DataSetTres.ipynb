Top 10 de productos con más porcentaje de tasa de pérdida por Item, para una mejor visualización
se construirá un tablero en Power BI

```python
PromedioTasaPerdida = round(DataSet3.groupby("Item Code")["Wholesale Price (RMB/kg)"].mean(), 2)
PromedioTasaPerdida.sort_values(ascending=False).head(10)
```

Top 10 de productos con más porcentaje de tasa de pérdida por Fecha, para una mejor visualización
se construirá un tablero en Power BI

```python
PromedioTasaPerdida2 = round(DataSet3.groupby("Date")["Wholesale Price (RMB/kg)"].mean(), 2)
PromedioTasaPerdida2.sort_values(ascending=False).head(10)
```

```python
# Tasa de Pérdida general
TasaPerdidaGeneral = DataSet3["Wholesale Price (RMB/kg)"].mean()
print(f"La tasa de pérdida general es de: % {round(TasaPerdidaGeneral, 2)}")
```
