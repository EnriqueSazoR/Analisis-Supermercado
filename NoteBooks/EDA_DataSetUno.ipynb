# Importación De Datos

```python
# DataSet Número 1
ruta = r"C:\Users\Luis Sazo\Desktop\Ventas Supermercados\Data\annex1.csv"
DataSet1 = pd.read_csv(ruta)
DataSet1
```

```python
# DataSet Número 4
ruta2= r"C:\Users\Luis Sazo\Desktop\Ventas Supermercados\Data\annex4.csv"
DataSet4 = pd.read_csv(ruta2)
DataSet4
```

Ambos DataSets se unirán mediante su Item Code, para que solo exista un DataSet de productos

```python
Productos = pd.merge(
    DataSet1, 
    DataSet4[["Item Code","Loss Rate (%)"]], 
    on = "Item Code")
Productos
```

```python
Productos.info()
```

```python
# Validar Nulos
Productos.isnull().sum()

# El DataFrame no presenta valores nulos
```

```python
# Promedio de tasa de perdida
promedioTasaPerdida = Productos["Loss Rate (%)"].mean()
print(f"La tasa de pérdida total es: % {round(promedioTasaPerdida, 2)}")
```

```python
AgruparCategorias = Productos.groupby("Category Name").count()
AgruparCategorias["Item Code"]
```

```python
# Exportar DataSet
Productos.to_csv("../Data_Limpia/Productos.csv", index=False)
```
